#### BeautifulSoup 패키지 다운 필요.

2) beatifulsoup으로 htmlparser

```python
import urllib.request
from bs4 import BeautifulSoup
```

daum에서 재무제표 정보 가져오기

```python
url = 'http://finance.daum.net/quote/all.daum?type=U&stype=P'
url2 = 'http://finance.daum.net/quote/all.daum?type=U&stype=Q'
```

```python
f = urllib.request.urlopen(url).read()
f2 = urllib.request.urlopen(url2).read()
```

```python
soup = BeautifulSoup(f, 'html.parser')
soup2 = BeautifulSoup(f2, 'html.parser')
```

```python
fields = soup.find_all('h4',{'class':'fl_le'}) 
fields2 = soup2.find_all('h4',{'class':'fl_le'})
```

```python
f = open('./data/kospi_code.txt', 'w')

for field in fields:
    
    # 업종명을 구합니다.
    fieldName = field.get_text().split('|')[0]
    
    # 현재 업종에 해당하는 종목들이 들어있는 'table'을 찾습니다.
    table = field.find_next_sibling('table') 

    # 'table'에서 속성 'class'의 속성값이 'txt'인 'td'를 모두 찾습니다.
    tds = table.find_all('td',{'class':'txt'})
    
    for td in tds:
        code = td.a['href'].split('=')[1]   # 종목코드를 구합니다.
        name = td.get_text()                # 종목명을 구합니다.
        # 현재가를 구하기 위해 다음 'td'를 찾습니다
        nextTd = td.find_next_sibling('td')  
        price = nextTd.get_text()           # 현재가를 구합니다.
 
        print(fieldName, code, name, price)
        f.write(code+'\n')
        
f.close()
```
